````python
# creating a series from a list

# To convert the list 'arr' into a pandas Series, we use the 'pd.Series()' function
# This function takes 'arr' as an argument and creates a Series object 's1' from it

# Indices: Automatically generated integer indices starting from 0 (unless specified otherwise)
# Values: Corresponding values from the arr list

# Each line shows an index (left-hand side) and its corresponding value (right-hand side)
# dtype: 'int64' indicates that the Series contains integers ('int64')

arr = [0, 1, 2, 3, 4]
s1 = pd.Series(arr)
s1
````

````python
# 'pd.Series(arr, index=order)' creates a Series 's2' where:
# values are taken from 'arr'
# indices are taken from 'order'

order = [1, 2, 3, 4, 5]
s2 = pd.Series(arr, index = order)
s2
````

````python
zz = [11, 22, 33, 44, 55, 66]          # list of values
# ind = [10, 20, 30, 40, 50, 60]
ind = ['a', 'b', 'c', 'd', 'e', 'f']   # list of custom indices (index has to be equal to the number of data in the series)
zz1 = pd.Series(zz, index = ind)       # values are taken from 'zz', indices are taken from 'ind'
zz1
````

````python
n = np.random.rand(5)                 # generates an array 'n' of 5 random numbers between 0 and 1
ind = ['a', 'b', 'c', 'd', 'e']       # list of custom indices (index has to be equal to the number of data in the series)
n1 = pd.Series(n, index = ind)        # values are taken from 'n', indices are taken from 'ind'
n1
````

***ARITHMETIC OPERATINS***

---




To add, subtract, multiply & divide two pandas Series 's3' and 's4', you can use the 'add', 'subtract', 'multiply' & 'divide' method or the '+', '-' , '*' & '/' operator. This operation performs element-wise addition, subtraction, multiplication and division aligning by the index. If an index is present in one Series but not the other, the resulting value for that index will be NaN.
```
s3 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s4 = pd.Series([4, 5, 6], index=['a', 'b', 'd'])

result = s3.add(s4)
result = s3.subtract(s4)
result = s3.multiply(s4)
result = s3.divide(s4)

print(result)
```
The output will be:
```
Addition -->
a    5.0
b    7.0
c    NaN
d    NaN
dtype: float64

Subtraction -->
a   -3.0
b   -3.0
c    NaN
d    NaN
dtype: float64

Multiplication -->
a     4.0
b    10.0
c     NaN
d     NaN
dtype: float64

Division -->
a    0.25
b    0.40
c     NaN
d     NaN
dtype: float64

```
Using the 'add' method provides more flexibility, such as allowing you to specify a 'fill_value' for missing values:
```
result = s3.add(s4, fill_value=0)
result = s3.subtract(s4, fill_value=0)
result = s3.multiply(s4, fill_value=0)
result = s3.divide(s4, fill_value=0)

print(result)
```
The output will be:
```
Addition -->
a    5.0
b    7.0
c    3.0
d    6.0
dtype: float64

Subtraction -->
a   -3.0
b   -3.0
c    3.0
d   -6.0
dtype: float64

Multiplication -->
a     4.0
b    10.0
c     3.0
d     6.0
dtype: float64

Division -->
a    0.25
b    0.40
c    3.00
d    0.17
dtype: float64
```
The 'fill_value' parameter is used to fill missing values with a specified value (in this case, '0').

Indices 'c' and 'd', which are only present in one of the Series, are treated as if the missing Series had a value of '0' at those indices.

````python
import matplotlib.pyplot as plt

# series and dataframe line chart

%matplotlib inline

ts = pd.Series(np.random.randn(100), index=pd.date_range('today', periods=100))     # create a time series with random values
ts = ts.cumsum()      # computes the cumulative sum of the series
ts.plot()             # plots the time series
````

````python
# 'index=ts.index' : Sets the index of the DataFrame to be the same as the index of the previously created time series 'ts',
# which is a range of 100 consecutive dates starting from today
# 'columns=['A', 'B', 'C', 'D']' : Sets the column names of the DataFrame to 'A', 'B', 'C', and 'D'

df = pd.DataFrame(np.random.randn(100, 4), index=ts.index, columns=['A', 'B', 'C', 'D'])

df = df.cumsum()        # computes the cumulative sum of the series
plt.figure()            # creates a new figure
df.plot()               # plots the DataFrame
plt.legend(loc='best')  # adds a legend to the plot, automatically determining the best location to place it to avoid overlapping with the plot lines


# the output is the plot of the cumulative sum of random values for four columns ('A', 'B', 'C', 'D') over 100 days, starting from today
# each line in the plot represents one of the columns from the DataFrame
# the x-axis shows the dates, while the y-axis shows the cumulative sum of the random values
# the legend indicates which color corresponds to each column, and it is placed in the best location to avoid overlapping with the lines
````

***REMOVE REPEATED VALUE***

---

````python
remove_data = pd.DataFrame({'A': [1, 2, 2, 2, 4, 4, 5, 5, 6, 7, 7]})    # creates a DataFrame named 'remove_data' with a single column 'A'
print("Original DataFrame:\n", remove_data)
print()

# 'drop_duplicates()' function removes all rows where all elements are the same as in a previous row
print("After removing the duplicate values (Using 'drop_duplicates()'):\n", remove_data.drop_duplicates())
print()

# 'shift()' function shifts the column values by one position, and 'loc' is used to keep only rows where the current value is different from the previous value
print("After removing the duplicate values (Using 'shift()'):\n", remove_data.loc[remove_data['A'].shift() != remove_data['A']])
````
