```python
fig = plt.figure()  # creates a new figure, a figure in matplotlib is essentially the entire window or page that will contain your plot(s)

# "fig.add_axes([left, bottom, width, height])": Adds a set of axes to the figure. The list [0.1, 0.1, 0.8, 0.8] specifies the position and size of the axes in the figure
# 'left': The distance from the left edge of the figure to the left edge of the axes (0.1 means 10% of the figure width)
# 'bottom': The distance from the bottom edge of the figure to the bottom edge of the axes (0.1 means 10% of the figure height)
# 'width': The width of the axes (0.8 means 80% of the figure width)
# 'height': The height of the axes (0.8 means 80% of the figure height)

axis = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)


axis.plot(x, y, 'r')      # plot the values of 'x' against 'y' with a red line ('r' stands for red color)
axis.set_xlabel('x')      # string that will be displayed as the label of the x-axis
axis.set_ylabel('y')      # string that will be displayed as the label of the y-axis
axis.set_title('title')   # string that will be displayed as the title of the plot, it helps to describe what the plot is about

plt.show()                # displays the plot
```

***fig, axis = plt.subplots(nrows=1, ncols=2)***

Return Value:

* ***fig*** : This is the Figure object. The Figure is the overall container for the plots. It holds all the Axes objects and their elements like titles, labels, and lines

* ***axis*** : This is an array-like object containing the Axes objects. The Axes are the individual plots where you can add data, labels, titles, and other elements. Since ncols=2, axis will contain two Axes objects

* ***'plt.subplots(nrows=1, ncols=2)'*** creates a new figure ('fig') and a set of subplots ('axis')

* ***'plt.subplots'*** --> matplotlib function used to create a figure and a grid of subplots

* ***'nrows=1'*** --> we want 1 row of subplots (if we think of it as a grid, we have one horizontal row)

* ***'ncols=2'*** --> we want 2 columns of subplots (we will have two vertical columns next to each other)

* together, 'nrows=1' and 'ncols=2' create a grid with one row and two columns of subplots

Therefore, 'axis' will be an array-like object containing the two subplot axes ('axis[0]' and 'axis[1]' in this case)

```python
fig, axis = plt.subplots(nrows=1, ncols=2)
for ax in axis:
  ax.plot(x, y, 'r')      # plot the values of 'x' against 'y' with a red line ('r' stands for red color)
  ax.set_xlabel('x')      # string that will be displayed as the label of the x-axis
  ax.set_ylabel('y')      # string that will be displayed as the label of the y-axis
  ax.set_title('title')   # string that will be displayed as the title of the plot, it helps to describe what the plot is about
```

```python
# draw contour map

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025                       # set the resolution of the contour map
xx = np.arange(-3.0, 3.0, delta)    # creates an array of 100 evenly spaced values from -3 to 3
yy = np.arange(-2.0, 2.0, delta)    # creates an array of 100 evenly spaced values from -2 to 2

X, Y = np.meshgrid(xx, yy)          # creates a meshgrid of X and Y


z1 = np.exp(-X**2 - Y**2)              # creates a 2D array of z1
z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)  # creates a 2D array of z2
Z = (z1 - z2) * 2                      # creates a 2D array of Z

print(xx)
print()
print(yy)
print()

fig, ax = plt.subplots()                      # creates a figure and axes object
CS = ax.contour(X, Y, Z)                      # creates a contour plot of Z
ax.clabel(CS, inline=1, fontsize=10)          # adds labels to the contour plot
ax.set_title('Simplest default with labels')  # string that will be displayed as the title of the plot, it helps to describe what the plot is about
```

```python
# draw 3D surface image
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 5))  # figsize=(10, 5) sets the size of the figure to 10 inches wide and 5 inches tall

# specify the 3D graphics to draw, with projection = '3D'
ax = fig.add_subplot(1, 2, 1, projection='3d')  # creates a subplot with 1 row and 2 columns, and set the current subplot to the first subplot

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, cmap=cm.coolwarm, linewidth=0, antialiased=False)  # plot the values of 'X', 'Y', 'Z' with a coolwarm color map
```

***plt.plot(x1, y1, marker='o', linestyle='-', color='b', label='sin(x1)'):***

* *plt.plot(...)* : This function creates a plot of y1 versus x1.

* *x1* : The data for the x-axis. This should be an array or list of x1-values.

* *y1* : The data for the y-axis. This should be an array or list of y1-values that correspond to the x1-values.

* *marker='o'* : Specifies that markers ('o' stands for circle) should be placed at each data point.

* *linestyle='-'* : Specifies that the line connecting the data points should be solid.

* *color='b'* : Specifies the color of the line and markers (in this case, 'b' stands for blue).

* *label='sin(x1)'* : Adds a label to the plot, which can be used in the legend to identify this line.

```python
# Generate x1 array
x1 = np.linspace(0, 10, 20)
print("First array x1:\n", x1)
print()

# Generate y1 array (let's use a simple function of x1 for illustration, e.g., y1 = sin(x1))
y1 = np.sin(x1)
print("Second array y1:\n", y1)
print()

# Plot x1 and y1
plt.plot(x1, y1, marker='o', linestyle='-', color='b', label='sin(x1)')  # plot the values of 'x1' against 'y1' with a blue line ('b' stands for blue color)
plt.title('Plot of sin(x1)')     # string that will be displayed as the title of the plot, it helps to describe what the plot is about
plt.xlabel('x1')                 # string that will be displayed as the label of the x-axis
plt.ylabel('y1')                 # string that will be displayed as the label of the y-axis
plt.legend()                     # displays a legend in the upper right corner of the plot
plt.grid(True)                   # adds a grid to the plot
plt.show()                       # displays the plot
```
